```python
cl1 = []
cl2 = []
n = n1 = n2 = 0

for i in s:
  n += 1
  if (i[0] < 1.25 and i[1]>6):
    cl1.append(i)
    n1 += 1
  else:
    cl2.append(i)
    n2 += 1
print(n1, n2)
print(n)
```

```python
def zov(spis):
  minp = []
  s = 0
  mins = float('inf')
  for i in spis:
    s = 0
    for j in spis:
      s += ((i[0]-j[0])**2 + (i[1]-j[1])**2)**0.5
    if s < mins:
      minp = i
      mins = s
  return minp
```

```python
z1 = zov(cl1)
z2 = zov(cl2)
```

```python
print((z1[0] + z2[0])*500)
print((z1[1] + z2[1])*500)
```
